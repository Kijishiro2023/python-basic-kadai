```python
import pandas as pd

list1 = pd.read_excel("2022_年間売上表.xlsx")
list2 = pd.read_excel("2023_年間売上表.xlsx")
print(list1)
print(list2)

df = pd.concat([list1, list2], ignore_index=True)
print(df)

result = df.groupby(["商品", "売上年"])["金額（千円）"].sum().reset_index()
print(result)

writer = pd.ExcelWriter("売上集計表.xlsx")
result.to_excel(writer, sheet_name="集計", index=False)
writer.close()

import openpyxl as ox
from openpyxl.styles import PatternFill
workbook = ox.load_workbook("売上集計表.xlsx")
sheet = workbook.active

fill = PatternFill(patternType="solid", fgColor="F2F2F2")

sheet["A1"].fill = fill
sheet["B1"].fill = fill
sheet["C1"].fill = fill

workbook.save("売上集計表.xlsx")
```
