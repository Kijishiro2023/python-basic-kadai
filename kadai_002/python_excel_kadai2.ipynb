```python
import pandas as pd
```

```python

# 売上データを読み込む
df = pd.read_excel('社員別売上データ.xlsx')
print(df)

# 売上の平均を求める
avg_sales = df['売上'].mean()

# 平均売上列を作成
df['平均売上'] = avg_sales
print(df)

#　売上平均との差で分岐し、業績ランクを求める関数「performance」を定義
def performance(sales):
  result = ''
  if sales >= avg_sales + 50 :
    result = 'A'
  elif sales >= avg_sales :
    result = 'B'
  else:
    result = 'C'
  return result

# 「業績ランク」列を作成し、関数「performance」を適用して値を設定
df['業績ランク'] = df['売上'].apply(performance)
print(df)

# Excelファイルを作成する
writer = pd.ExcelWriter('業績.xlsx')

# DataFrameオブジェクトをExcelファイルに書き込む
df.to_excel(writer, sheet_name='Sheet1', index=False)
writer.close()
```
