```python
#openpyxlをインポート
import openpyxl

workbook = openpyxl.Workbook()
sheet = workbook.active

# datetimeモジュールのdatetimeクラスをインポート
from datetime import date

# 現在の日付（年月日）を取得して、変数todayに代入
today = date.today()

#シートに帳票名と連絡先を追加
sheet['A1'].value = '請求書'
sheet['A2'].value = '株式会社ABC'
sheet['A3'].value = '〒101-0022 東京都千代田区神田練塀町300'
sheet['A4'].value = 'TEL:03-1234-5678 FAX:03-1234-5678'
sheet['A5'].value = '担当者名:鈴木一郎 様'

#シートにNoと日付欄を追加
sheet['E2'].value = 'No.'
sheet['E3'].value = '日付'
sheet['F2'].value = '0001'
sheet['F3'].value = today.strftime('%Y/%m/%d')


# 表のヘッダー行を書き込む
sheet.append(['商品名', '数量', '単価', '金額'])

# データを書き込む
items = [
    ['商品A', 2, 10000 ],
    ['商品B', 1, 15000 ],
]

# 金額を計算してから追加
for name, qty, price in items:
    sheet.append([name, qty, price, qty * price])

# 金額合計を追加
total = sum(qty * price for _, qty, price in items)
sheet['D9'] = total

# シートに小計と消費税、合計欄を追加
sheet['A10'].value = '小計'
sheet['A11'].value = '消費税'
sheet['A12'].value = '合計'

# 金額合計を追加
tax = total*0.1
all_total = total + tax
sheet['D10'].value = total
sheet['D11'].value = tax
sheet['D12'].value = all_total

# 行と列を追加し、レイアウトを調整
sheet.insert_rows(1, 1)
sheet.insert_rows(3, 1)
sheet.insert_rows(8, 2)
sheet.insert_rows(14, 1)
sheet.insert_cols(1, 1)

# 日付のtodayオブジェクトを、特定のフォーマットで文字列に変換する
filename = f"請求書_{today.strftime('%Y%m%d')}.xlsx"

workbook.save(filename)
```
