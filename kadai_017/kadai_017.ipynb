```python
class Human :
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult (self) :
        if self.age >= 20 :
             print(f"{self.name}は大人です。")
        else :
             print(f"{self.name}は大人ではありません。")

humanA = Human("じろう", 20)
humanB = Human("いちろう", 30)
humanC = Human("さぶろう", 19)

human_list = [humanA, humanB, humanC]

for human_check in human_list :
    human_check.check_adult()
```
