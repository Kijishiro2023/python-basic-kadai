```python
# 購入済みかどうかを判定する変数
purchased = True

# 戻り値を返す関数を定義する
def is_purchased():
    if purchased:
        return True
    else:
        return False

def add_two_arguments(price, tax):
    # 与えられた引数priceと引数taxを加算し、変数totalに代入する
    total = price + (price * tax / 100 )
    return total   # ← 戻り値を返す

# 関数を呼び出し、引数として購入金額と消費税を渡す
total = add_two_arguments(110, 10)  # ← 戻り値を受け取る

# 戻り値を返す関数を条件式に使う（Trueであれば処理が実行される）
if is_purchased() == True:
    print(f"{total}円")
```
